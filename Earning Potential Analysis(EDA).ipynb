```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('adult_data.csv', names=['age','workclass', 'fnlwgt','education', 'education-num', 'marital-status','occupation','relationship','race','sex',
                                   'capital-gain','capital-loss','hours-per-week','native-country','earning'])
data
```

```python
data.drop(columns=['capital-gain','capital-loss'],inplace=True)
```

```python
data.replace(' ?',np.nan, inplace=True)
data.dropna(axis=0, inplace=True)
```

```python
data.reset_index(inplace=True, drop=True)
```

```python
missing_values=[' ? ', '  ',' NA ']
for i in missing_values:
   data.replace(i, np.nan, inplace=True)
```

```python
data['Experience']=data['age']-data['education-num']-5
data
```

```python
data['Work-life Balance']=data['hours-per-week']/data['age']
data
```

```python
data_clean=data.copy()
num_cols=['age','education-num','hours-per-week']
for col in num_cols:
   q1=data_clean[col].quantile(0.25)
   q3=data_clean[col].quantile(0.75)
   iqr=q3-q1
   lower_limit=q1-1.5*iqr
   upper_limit=q3+1.5*iqr
   data_clean=data_clean[(data_clean[col]>=lower_limit) & (data_clean[col]<=upper_limit)]
data_clean.shape
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='earning',y='hours-per-week',data=data_clean)
```

```python
data_clean['earning_potential']=data_clean['earning'].apply(lambda x:1 if '>50K' in x else 0)
corr=data_clean.corr(numeric_only=True)['earning_potential']
print(corr)
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
for col in data_clean.columns:
   if data_clean[col].dtype=='object':
       data_clean[col]=label.fit_transform(data_clean[col])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.linear_model import LogisticRegression
```

```python
X=data_clean.drop(['earning','earning_potential'],axis=1)
y=data_clean['earning']
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
X_scaled=s.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test=train_test_split(X_scaled,y,test_size=0.2, random_state=1)
model=LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
print('Accuracy is: ',accuracy_score(y_test,y_pred)*100)
print('Precision: ',precision_score(y_test,y_pred))
print('Recall: ',recall_score(y_test,y_pred))
print('F1 score: ',f1_score(y_test,y_pred))
```
